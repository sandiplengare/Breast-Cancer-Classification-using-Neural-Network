## Exploratory Data Analysis (EDA)

- Exploratory Data Analysis is all about analyzing the dataset and summarizing the key insights and characteristics of the data.
- EDA Checklist
  - 1. Understanding the dataset, and its shape.
  - 2. Checking the datatpye of each feature.
  - 3. Categorical and Numerical features.
  - 4. Checking for missing values.
  - 5. Descriptive summary of the dataset.
  - 6. Groupby for classification problem.

## Importing the Dependencieas

```python
import numpy as np                              # To perform numerical operations
import pandas as pd                             # To perform data processing
from sklearn.preprocessing import LabelEncoder  # To convert categorical feature into numerical feature
```

## Data Collection

```python
# Loading the dataset

breast_cancer_data = pd.read_csv('data.csv')
```

## Exploratory Data Analysis

```python
# Printing the first five rows of the dataframe

breast_cancer_data.head()
```

```python
# Removing the unnamed column

breast_cancer_data.drop(['Unnamed: 32'], axis = 1, inplace=True)
```

```python
breast_cancer_data.sample()
```

```python
# Checking the shape of the dataframe

breast_cancer_data.shape
```

```python
# Checking the datatypes

breast_cancer_data.info()
```

```python
# Removing ID column

breast_cancer_data.drop(['id'], axis=1, inplace=True)
```

- Diagnosis column is a categorical column whereas remaining are contineous columns.

```python
# Checking for missing values

breast_cancer_data.isnull().sum()
```

- There is no missing values in the dataset.

```python
# Statistical summary of the data - Descriptive statistics

breast_cancer_data.describe()
```

- Check whether mean and median are colse to each other.

```python
# Checking the distribution of target variable

breast_cancer_data['diagnosis'].value_counts()
```

```python
# Encoding the target column

label_encode = LabelEncoder()

labels = label_encode.fit_transform(breast_cancer_data['diagnosis'])

breast_cancer_data['target'] = labels

breast_cancer_data.drop(['diagnosis'], axis=1,inplace=True)
```

```python
breast_cancer_data.sample()
```

```python
breast_cancer_data['target'].value_counts()
```

- Benign --> 0

- Malignant --> 1

There is a slight imbalance in the data. But it is fine in this case

```python
# Grouping the data based on the target

breast_cancer_data.groupby('target').mean()
```

- Inference: We can clearly see that for most of the features, the mean values are higher for Malignant(1) cases and lower for Benign(0) cases

## Summary from EDA

- No missing values
- All are continuous numerical values except for target column.
- Mean is slightly more than the medianfor most of the features. So it is skewed.
- Slight imbalnce in the data.
- Mean of most features are larger for Malignant cases compared to the Benign cases.
